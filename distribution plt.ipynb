

# The distribution plot is suitable for comparing range and distribution for groups of numerical data
#When to use it
The distribution plot is suitable for comparing range and distribution for groups of numerical data.

#Advantages
Distribution plot organizes large amount data and visualizes outlier values 

#Disadvantages
The distribution plot is not relevant for detailed analysis of the data as it deals with a summary of the data distribution.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

%matplotlib inline
```

```python
num=np.random.randn(150)
```

```python
sn.distplot(num)
```

```python
sn.distplot(num,color='green')
```

```python
label_dist=pd.Series(num,name="variable x")
```

```python
sn.distplot(label_dist)
```

```python
sn.distplot(label_dist,vertical=True,color='green')
```

```python
sn.distplot(label_dist,hist=False)
```

# density plot

```python
A density plot is a representation of the distribution of a numeric variable.
 A Density Plot visualises the distribution of data over a continuous interval or time period.
```
